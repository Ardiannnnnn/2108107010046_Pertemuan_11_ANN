```python
import numpy as np # type: ignore
import matplotlib.pyplot as plt # type: ignore
import pandas as pd # type: ignore
from sklearn.preprocessing import LabelEncoder, OneHotEncoder # type: ignore
from sklearn.compose import ColumnTransformer # type: ignore
from sklearn.model_selection import train_test_split # type: ignore
from sklearn.preprocessing import StandardScaler # type: ignore
from tensorflow.keras.models import Sequential # type: ignore
from tensorflow.keras.layers import Dense # type: ignore
from sklearn.model_selection import train_test_split # type: ignore
import pandas as pd # type: ignore
from sklearn.metrics import classification_report # type: ignore
```

```python
dataset = pd.read_csv('Obesity Classification.csv')
```

```python
dataset['BMI'] = round(dataset['BMI']).astype(int)
```

```python
dataset.info()
```

```python
X = dataset.iloc[:, 3:6].values
y = dataset.iloc[:, 6].values
```

```python
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)

print(y)
```

```python
# Membagi data ke test dan training set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
# Feature Scaling
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Inisialisasi ANN
MesinKlasifikasi = Sequential()

# Menambah input layer dan hidden layer pertama
MesinKlasifikasi.add(Dense(units=12, kernel_initializer='uniform', activation='relu', input_dim=3))

# Menambah hidden layer kedua
MesinKlasifikasi.add(Dense(units=6, kernel_initializer='uniform', activation='relu'))

# Menambah output layer dengan activation='softmax' untuk klasifikasi multi-kelas
MesinKlasifikasi.add(Dense(units=4, kernel_initializer='uniform', activation='softmax'))

# Compile model dengan loss='sparse_categorical_crossentropy' untuk klasifikasi multi-kelas
MesinKlasifikasi.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Fitting ANN ke training set
history = MesinKlasifikasi.fit(X_train, y_train, batch_size=10, epochs=100, validation_data=(X_test, y_test))
```

```python
# Melakukan prediksi pada set pengujian
y_pred = MesinKlasifikasi.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)

# Mengonversi label yang sudah diencode kembali ke label aslinya
y_test_orig = label_encoder.inverse_transform(y_test)
y_pred_orig = label_encoder.inverse_transform(y_pred_classes)

# Mencetak confusion matrix
cm = confusion_matrix(y_test_orig, y_pred_orig)
print("Confusion Matrix:")
print(cm)
```
